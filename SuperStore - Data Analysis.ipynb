One common issue that arises in datasets is missing values, which can affect the accuracy of our analysis. In this case, we are specifically concerned with missing values in the "Postal_Code" column. To replace null values in the postal code column of a DataFrame based on the city, we can create a dictionary mapping cities to their postal codes. For each null value in the postal code column, we will look up the corresponding city and replace the null value with the postal code of that city. This approach ensures that each city's postal code is correctly assigned even if some postal codes are missing in the original dataset.

1. **Dictionary Creation:** 
   - Create a dictionary mapping each city to its postal code.

2. **Replace Nulls:**
   - For each row with a null postal code, use the city value to look up and replace the null with the corresponding postal code from the dictionary.

3. **Considerations:**
   - Ensure all unique cities are covered in the dictionary.
   - Handle cases where a city has no corresponding postal code.
   - Validate the correctness and consistency of the replacements.

```python
def create_city_to_postal_dict(df, city_col, postal_col):
    # Initialize an empty dictionary to store city-postal code mappings
    city_to_postal = {}
    
    # Create a new DataFrame containing unique city-postal code pairs
    city_postal_pairs = df[[city_col, postal_col]]
    
    # Iterate over each row in the DataFrame
    for index, row in city_postal_pairs.iterrows():
        # Check if the postal code is not null
        if pd.notnull(row[postal_col]):
            # Add the city-postal code pair to the dictionary
            city_to_postal[row[city_col]] = row[postal_col]
    
    # Return the dictionary mapping cities to postal codes
    return city_to_postal

# Function to replace null postal codes based on city-postal code mappings
def replace_null_postal_codes(df, city_col, postal_col):
    city_to_postal = create_city_to_postal_dict(df, city_col, postal_col)# Create city-postal code mappings
     # Use apply function to replace null postal codes
    df[postal_col] = df.apply(lambda row: city_to_postal[row[city_col]] if pd.isnull(row[postal_col]) and row[city_col] in city_to_postal else row[postal_col], axis=1)
    return df

# Replace null postal codes
df = replace_null_postal_codes(df, 'City', 'Postal_Code')
df.isna().sum()
```

3. <u><h3>Clustering Customers based on Region, Category, Segment and Shipment Mode</h3></u>
- This code clusters customers based on their region, category, segment, and shipment mode preferences. 
- It uses clustering algorithms to group customers with similar preferences together.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

def create_cluster_pivot_table(df, column_name):
    # Selecting relevant columns for clustering
    customer_data = df.groupby('Customer_ID')[column_name].agg(lambda x: x.value_counts().index[0]).reset_index()

    # Encoding categorical variables
    customer_data_encoded = pd.get_dummies(customer_data, columns=[column_name])

    # Standardizing the data
    scaler = StandardScaler()
    customer_data_scaled = scaler.fit_transform(customer_data_encoded.drop('Customer_ID', axis=1))

    # Clustering using KMeans
    kmeans = KMeans(n_clusters=4, random_state=0)
    customer_data['Cluster'] = kmeans.fit_predict(customer_data_scaled)

    # Creating a pivot table for displaying customer IDs under each cluster
    pivot_table = customer_data.pivot_table(index='Customer_ID', columns='Cluster', aggfunc=lambda x: 'X').fillna('')
    
    return pivot_table

# Assuming your DataFrame is named 'superstore_df'
# Replace 'superstore_df' with the actual name of your DataFrame if it's different

# Create pivot tables for Region, Sub_Category, and Segment
pivot_table_region = create_cluster_pivot_table(df, 'Region')
pivot_table_category = create_cluster_pivot_table(df, 'Category')
pivot_table_segment = create_cluster_pivot_table(df, 'Segment')
pivot_table_ship_mode = create_cluster_pivot_table(df, 'Ship_Mode')
```

5. <u><h3>Classification Analysis</h3></u>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Define the classification tasks
criterion = {
    'Region': ['State'],
    'Category': ['Sub_Category']
}

def perform_classification(df, X_cols, y_col):
    # Selecting relevant features for classification
    X = df[X_cols]
    y = df[y_col]

    # Encoding categorical variables
    X_encoded = pd.get_dummies(X, columns=X_cols)

    # Splitting the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=0)

    # Creating and training the Random Forest Classifier model
    rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)
    rf_classifier.fit(X_train, y_train)

    # Making predictions
    y_pred = rf_classifier.predict(X_test)

    # Evaluating the model
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Perform classification for each criterion
for label, cols in criterion.items():
    print(f"\nPerforming {label} Classification")
    perform_classification(df, cols, label)
```

6. <h3><u>Customer Lifetime Value Analysis (CLTV)</u></h3>
- The code calculates the Customer Lifetime Value (CLTV) for each customer.
- It does this by first grouping the data by 'Customer_ID' to calculate the total sales, number of unique orders, and average order value for each customer.
- Then, it calculates the average purchase frequency rate across all customers.
- Finally, it computes the CLTV by multiplying the average order value by the average purchase frequency rate.

```python
# Assuming 'Customer_ID' is a unique identifier for each customer
# Calculate total sales for each customer
customer_total_sales = df.groupby('Customer_ID')['Sales'].sum().reset_index()

# Calculate the number of unique orders for each customer
customer_order_count = df.groupby('Customer_ID')['Order_ID'].nunique().reset_index()

# Calculate average order value for each customer
customer_avg_order_value = customer_total_sales['Sales'] / customer_order_count['Order_ID']

# Calculate the average purchase frequency rate
purchase_frequency = len(df) / df['Customer_ID'].nunique()

# Calculate customer lifetime value
customer_lifetime_value = customer_avg_order_value * purchase_frequency

# Create a DataFrame with Customer_ID and CLTV
cltv_df = pd.DataFrame({'Customer_ID': customer_total_sales['Customer_ID'], 'CLTV': customer_lifetime_value})

# Display the CLTV for each customer
cltv_df
```

7. <h3><u>Churn Analysis</u></h3>
- The code analyzes customer churn rate and factors influencing it.
- It calculates the churn rate by first identifying the latest order date for each customer and then calculating the total number of orders and the average number of days between orders.
- Customers who haven't made a purchase in the last 365 days are considered churned.
- The churn rate is then calculated as the ratio of churned customers to the total number of customers.

```python
# Assuming 'Churn' is a binary column indicating whether a customer has churned (1) or not (0)
# 'Order_Date' is the column indicating the date of the order

# Calculate the latest order date for each customer
customer_latest_order_date = df.groupby('Customer_ID')['Order_Date'].max().reset_index()

# Calculate the total number of orders for each customer
customer_order_count = df.groupby('Customer_ID')['Order_ID'].nunique().reset_index()

# Calculate the total number of days between the latest order date and today's date
current_date = pd.to_datetime('today')
customer_latest_order_date['Days_since_last_order'] = (current_date - customer_latest_order_date['Order_Date']).dt.days

# Calculate the average number of days between orders for each customer
customer_avg_days_between_orders = customer_latest_order_date['Days_since_last_order'] / customer_order_count['Order_ID']

# Calculate churn rate
churn_rate = sum(customer_latest_order_date['Days_since_last_order'] < 365) / len(customer_latest_order_date)

# Display churn rate
print("Churn Rate:", churn_rate)
```

- A churn rate of 0.0 means that none of the customers have churned. 
- In other words, all customers are still active and have made a purchase within the specified time frame. 
- This is a positive sign for the business, indicating high customer retention and loyalty.<br>
<p>However, it's important to note that the calculation of churn rate depends on the time frame and definition of churn used, so a churn rate of 0.0 should be interpreted in the context of these factors.

8. <h3><u>Market Basket Analysis using Association Rules:</u></h3><br>
- The association rule mining analysis was conducted to identify patterns in customer purchasing behavior. 
- The results reveal strong associations between different segments, categories, and sub-categories of products. 

```python
# Importing required libraries
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Loading the dataset
df = pd.read_csv("superstore_ds.csv")
df

# Convert categorical data to one-hot encoding
df_encoded = pd.get_dummies(df[['Segment', 'Category', 'Sub_Category']], drop_first=True)

# Find frequent itemsets
frequent_itemsets = apriori(df_encoded, min_support=0.005, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Display the association rules
rules
```

These associations can be summarized as follows:

1. **Segment-Category Association:**
   - Customers in the Corporate and Home Office segments are strongly associated with purchasing Office Supplies.
   - The Corporate segment also shows a strong association with purchasing Technology products.

2. **Segment-Sub-Category Association:**
   - Several sub-categories such as Appliances, Art, Binders, Envelopes, Fasteners, Labels, Paper, Storage, Supplies, and Phones show a perfect association with the Corporate and Home Office segments.
   - This indicates that customers in these segments tend to purchase these sub-categories exclusively.

3. **Category-Sub-Category Association:**
   - Office Supplies are associated with various sub-categories, indicating that customers purchasing Office Supplies also tend to buy these specific product types.

These associations provide valuable insights for marketing and product placement strategies. For example, targeting promotions for Technology products towards Corporate customers and optimizing product bundling for Office Supplies and related sub-categories can potentially increase sales and customer satisfaction.

It's important to note that association rules are based on historical transaction data and should be used as a guideline rather than a definitive rule for customer behavior.
